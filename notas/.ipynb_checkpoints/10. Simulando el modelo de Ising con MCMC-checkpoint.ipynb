### 1.-

Sabemos que si $\pi(\ss)$ es la probabilidad estacionaria de estar en el lugar $\ss$, entonces la probabilidad de aceptaci\'on de pasar a $\tt$ es
\begin{equation}
\alpha(\ss\rightarrow\tt)=\frac{\pi(\tt)}{\pi(\ss)}
\end{equation}
pero $\pi(\ss)=e^{-\beta E(\ss)}$por lo tanto
\begin{equation}
\alpha(\ss\rightarrow\tt)=e^{-\beta \Delta E}
\end{equation}


### 2.- 

La energía del estado $\ss$ está dada por:
\begin{equation}
E(\ss)=-\sum_{\langle i,j\rangle} J_{i,j}\sigma_i \sigma_j
\end{equation}
con $i,j$ vecinos en la red de espines. Para simplificar el  calculo asumimos que $J_{i,j}=1\ \ \forall i,j$.
Si volteamos el espín i' entonces, cuatro de los elementos de la suma cambian de signo (los relacionados con los vecinos de $i'$), esto es
\begin{equation}
\Delta E=E(\tau)-E(\sigma)=-\sum_{\text{$j$ vc a i'}} \sigma_j(\tau_{i'}-\sigma_{i'})
\end{equation}
(vc se refiere a vecinos cercanos), pero $\tau_{i'}-\sigma_{i'}=-2\sigma_{i'}$, por lo tanto
\begin{equation}
\Delta E=E(\tau)-E(\sigma)=2\sigma_{i'}\sum_{\text{$j$ vc a i'}} \sigma_j
\end{equation}

```julia
#Temperatura y beta
#kB=1.3806488*10.0^(-23)
kB=1
β(T)=1/(kB*T);
T=0.2
β(T)
```

```julia
####Configuracion y funciones de flip,energia y magnetización####

#Configuracion aleatoria en un matriz
function config(N,M)
    2*int(rand(N,M))-1
end

#Hacemos nuestro propio modulo para tener fronteras periodicas
function modulo(i::Int64,L::Int64)
    i = i>L ? 1 : i<1 ? L : i
end

#Flip a la configuracion
function flip!(σ::Array{Int64,2},i::Int64=0,j::Int64=0)               
  N=size(σ)[1]
  M=size(σ)[2]
  if i==0 && j==0
    i,j=int((N-1)*rand())+1,int((M-1)*rand())+1
  end
  σ[i,j]=-σ[i,j]
  σ
end

#Calculo de Energia
function energia(σ)
    L1=size(σ)[1]
    L2=size(σ)[2]
    E=0
    for i in 1:L1
        for j in 1:L2
            E+=-σ[i,j]*(σ[modulo(i+1,L1),j]+σ[modulo(i-1,L1),j]+
                σ[i,modulo(j+1,L2)]+σ[i,modulo(j-1,L2)])
        end
    end
    E
end

function magnet(σ)
    sum(σ)
end

#Cambio de energia en un flip propuesto
function δE(σ::Array{Int64,2},i::Int64=0,j::Int64=0)
    if i==0 && j==0
    i,j=int((size(σ,1)-1)*rand())+1,int((size(σ,2)-1)*rand())+1
    end
    L1=size(σ)[1]
    L2=size(σ)[2]
    
    2*σ[i,j]*(σ[modulo(i+1,L1),j]+σ[modulo(i-1,L1),j]+
    σ[i,modulo(j+1,L2)]+σ[i,modulo(j-1,L2)]),
    i,j
end
```

```julia
#Proceso estocastico
function metropolis!(σ::Array{Int64,2},T::Float64)    
    ΔE,i,j=δE(σ)
    ΔM=0
    α=exp(-β(T)*ΔE)
    r=rand()
    cambio=false
    if r<α
        σ[i,j]=-σ[i,j]
        cambio=true
        ΔM=σ[i,j]<0?-2:2
    else
        σ[i,j]=σ[i,j]
        cambio=false
    end
    return cambio,ΔE,ΔM
end
```

### 4.-

```julia
######Evolucion de una cadena#######

function iteracion_con_E_y_M(espines::Array{Int64,2},iteraciones::Int64,T::Float64=30.0)
    Es::Array{Int64}=Array(Int64, iteraciones)
    Ms::Array{Int64}=Array(Int64, iteraciones)
    cambios::Array{Bool}=Array(Bool, iteraciones)
    E=energia(espines)
    M=magnet(espines)
    
    for i in 1:iteraciones
        cambio,ΔE,ΔM=metropolis!(espines,T)
        cambios[i]=cambio
        if cambio
            E+=ΔE
            M+=ΔM
        end
        Es[i]=E
        Ms[i]=M
    end
    
    return Es,Ms,cambios
end
```

```julia
#Vamos a definir una funcion q le dará formato a los ticks en x, que son muy grandes y se ven mal
function cientifico_corto_con_veces(x, p, veces)
    dim=veces/10
    magn=log10(dim)
    return string(int(x/(dim)))*"e"*string(int(magn))
end
```

```julia
#### Prueba de recorrido

#Valores iniciales
ancho=100
largo=100
espines=config(ancho,largo)
veces=int(5e6)
T=2.2

#Ploteo
plot(iteracion_con_E_y_M(espines,veces,T)[1])
```

```julia
##### Simulación para diferentes T ################

#Valores iniciales
ancho=100
largo=100
N=ancho*largo
veces=int(4e5)
espines=config(ancho,largo)
espinesInicial=espines
Ts=[10.0,8.0,6.0,4.,2.,1.]
Es=Array(Array{Float64,1},length(Ts))
Ms=Array(Array{Float64,1},length(Ts))
ds=Array(Array{Bool,1},length(Ts))

#Simulacion
for i in 1:length(Ts)
    espines=copy(espinesInicial)
    Es[i],Ms[i],ds[i]=iteracion_con_E_y_M(espines,veces,Ts[i])
end
```

```julia
###### Ploteo ########
cientifico_corto(x, p)=cientifico_corto_con_veces(x, p, length(Es[1]))
for j in 1:2
    #Estructura base
    fig,axs=subplots(2,3)
    fig[:subplots_adjust](hspace=0.3,wspace=0.3)
    fig[:set_size_inches](12,7)
    #Contador
    i= j==1 ? 0 : 6

    for ax in axs
        i+=1
        #Estilo
        ax[:xaxis][:set_tick_params](labelsize=10)
        ax[:yaxis][:set_tick_params](labelsize=10)
        ax[:xaxis][:set_label_text]("# de iteración")
        f = matplotlib[:ticker][:FuncFormatter](cientifico_corto) # Definimos el formato de tick labels
        ax[:xaxis][:set_major_formatter](f) # lo aplicamos
        #Ploteo
        if i%2!=0
            ax[:plot](Es[(i+1)/2])
            ax[:set_title]("T=" * @sprintf("%1.2f",(Ts[(i+1)/2])))
        else
            ax[:plot](Ms[(i)/2])
        end
    end

    #Estilo especifico
    axs[1][:yaxis][:set_label_text]("Energía")
    axs[2][:yaxis][:set_label_text]("Magnetización")
    fig[:suptitle]("Energia y Magnetizacion en Ising")
end
```


### 5.-

Haciendo pruebas con diferentes tamaños de configuración, pudimos observar que despues de aproximadamente $N$ (con $N$ el numero de estados de la conf) iteraciones, se consigue el comportamiento que se espera de equilibrio térmico.

Esto tiene sentido pues es cuando se ha dado la oportunidad de recorrer al menos una vez la matriz entera. Es por eso que decidimos tomar un promedio de los valores despues de $N$ iteraciones. A esto se le llama el perido de quemado (burn-in period).

Debido a que este es un proceso de markov que se basa en el anterior y que se mueve solo espin por espin, la correlacion en sampleos muy pequeños es grande, para evitar esto tomaremos el promedio con medidas cada $\sqrt N$ iteraciones.

```julia
function promedio_propiedad(propiedad::Array{Float64,1},estados::Int64,funcion_tamano_sampleo::Function=sqrt)
    sumProp::Float64=0
    medidas::Int64=0
    for i in 1:int(funcion_tamano_sampleo(estados)):length(propiedad)
        medidas+=1
        sumProp+=propiedad[i]
    end
    sumProp/medidas
end
```

```julia
##### Calculo del promedio de E para diferntes βs #####

#Unos atajitos
ts_num=length(Ts)
E_proms=Array(Float64,ts_num)

#Sacamos el promedio
for i in 1:ts_num
    E_proms[i] = promedio_propiedad(Es[i],veces)
end

#pasamos de T a beta
beta=similar(Ts)
for i in 1:ts_num
    beta[i]=1/Ts[i]
end

#Ploteamos
plot(beta,E_proms,"rx-")
#xscale("log")
#yscale("log")
#ax=gca()
#ax[:set_xlim]([0,0.2])
#ax[:set_ylim]([-4000,0])
#ax[:xaxis][:zoom](1)
xlabel(L"β")
ylabel(L"⟨E⟩_β")
```
