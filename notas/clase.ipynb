```python
function f(p,n)
    x=[0]
    for a=[1:n]
        if rand()<=p
            push!(x,x[a]+1)        
        else
            push!(x,x[a]-1)
        end
    end
    return x
end
```

```python
plot(f(0.55,1000))
```

```python
function caminantes(p,n,m)
    xnew=zeros(m)
    for a=[1:m]
        y=f(p,n)
        yn=y[n]
        xnew[a]=yn
    end
return xnew
end        
```

```python
caminantes(0.5,100000,100)
```
