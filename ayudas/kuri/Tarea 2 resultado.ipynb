Considérese un caminante aleatorio que vive en los enteros. Es simétrico, es decir, en cada paso, brinca a la izquierda o al derecha, los dos con probabilidad $\frac{1}{2}$. Empieza en $1$. 

Estudiaremos el **tiempo de primer paso** a $0$, es decir, ¿cuándo llegará por primera vez el caminante a $0$? A esta variable aleatoria la llamaremos $\tau$.

## Preguntas

[1] Escribe una función que simule al caminante aleatorio, empezando en $1$ hasta que llegue a $0$. El caminante vive en una caja de tamaño $L+1$, es decir en $\{0, 1, \ldots, L-1, L\}$, y la función debe tomar $L$ como parámetro.

Hay una *frontera reflejante* en $L+\frac{1}{2}$: si el caminante está en $L$ e intenta brincar a la derecha, se refleja y se queda en donde estaba.

La función regresa el tiempo que tardó en llegar a $0$.

[2] Escribe una función que calcule el tiempo promedio $\langle \tau_L \rangle$ en llegar a $0$ para un sistema de tamaño $L$, sobre un número de corridas $N$.

[3] Dibuja el tiempo promedio $\langle \tau_L \rangle$ en función del tamaño $L$. ¿Cómo escala el tiempo promedio con $L$? ¿Qué pasa cuando $L \to \infty$?

[4] Escribe una función que calcule la distribución de probabilidad de $\tau_L$. 

[5] Dibuja la distribución de probabilidad $f_{\tau_L}(t)$ de $\tau_L$ para distintas $L$ en una sola gráfica. ¿Cómo escala la cola de la distribución? 

[6] ¿Cómo se ve la distribución de $\tau_\infty$? ¿Cómo es su cola? ¿Qué implica para el tiempo promedio de $\tau_\infty$?

Respuesta inciso 1)

```julia
function caminante(L) #L es la longitud de la caja
    pos=1
    t=0
    while pos > 0     # lo hace mientras no lleguemos al 0
        t+=1
        aux=rand()
        if aux>0.5     #si el rand da mayor que 0.5, va a la derecha, checando que aun haya espacio para ir a la derecha
            if pos < L
                pos+=1
            end
        else           #si no, a la izquierda
            pos-=1
        end
    end
    return t
end

x=caminante(10)
```

Respuesta inciso 2)

```julia
function cami_prom(L,N) #Usamos la misma función, N es el número de corridas
    t_prom=0.0
    for i in 1:N            
        t_prom+=caminante(L)
    end
    t_prom=t_prom/N  #Sumando ya todos los valores, ahora sí dividimos entre el número total de intentos
end

y=cami_prom(1000,100000)
```

Respuesta inciso 3)

```julia
L_Max=500   #Longitud máxima de la caja
N=10000    #numero de corridas
promedios=zeros(L_Max)

for i in 1:L_Max
    promedios[i]=cami_prom(i,N)
end

u=[1:L_Max]

plot(u,promedios,"o")
```

De la gráfica notamos que $\tau_L$ escala como $2L$, Por lo que cuando L tiende a infinito, $\tau_L$ también

Respuesta inciso 4)

```julia
function dist_caminante(L,N) #N será el número de intentos
    distri=zeros(10*L) 
    for i in 1:N            
        t=caminante(L)
        if t<=length(distri) #la longitud del array resulta un problema
            distri[t]+=1     #si el numero de pasos es menor a la longitud de donde estamos guadando la distribucion, no pasa nada
        else                 #si es mayor, extiende el array
            f=t-length(distri)
            extension=zeros(f)
            distri=vcat(distri,extension)
            distri[t]+=1 
        end
    end
    distri=float(distri)/float(N)
    u=[1:length(distri)]
    plot(u,distri,"o")
end

x=dist_caminante(10,1000)
```

Respuesta inciso 5)

```julia
function dist_caminante2(L,N,len,zoom) #vamos a cambiar cómo se maneja la longitud del array
    distri=zeros(int(len/2)+2)         #len es la longitud del array que vamos a capturar
    num_excesos=0                      #las veces que la cantidad de pasos se pasan de la longitud del array
    for i in 1:N                       #hacemos el array de distribucion de poco mas de la mitad de largo, ya que todos los 
        t=caminante(L)                 #lugares pares seran 0
        if t<=len && distri[int((t+1)/2)] < N/zoom     #el zoom es para delimitar el máximo en el eje y, entre mayor zoom      
            distri[int((t+1)/2)]+=1                    #el maximo del eje y es menor
        else                            #si se pasa, no lo registramos, y sumamos 1 a lo que le vamos a restar a N, porque se
            num_excesos+=1              #promediara sobre menos outputs  
        end
    end
    N=N-num_excesos
    distri=float(distri)/float(N)
    u=zeros(int(len/2)+2)
    for k in 1:length(u)
        u[k]=2*k-1
    end
    plot(u,distri)
end

#for j in 1:5
 #   x=dist_caminante2(j*20,1000000,400)
#end

x=dist_caminante2(10,1000000,1000,400)
x=dist_caminante2(100,1000000,1000,400)
```

De la grafica anterior, la azul es la primera, o sea de L=10 es la azul, la de L=100 es la verde, notamos como para mayor L, a valores pequeños la distribución baja más abruptamente que la de menor L, pero para mayor L, también es más probable un mayor numero de pasos, representados en que la distribucion de L=100, va por arriba que la de L=10

Respuesta inciso 6)

la distribucion de $\tau_\infty$ empieza en 1 alrededor de 0.5, y de ahí todos los demás valores tienden a 0, pero para cantidad de pasos grande, la probabilidad de $\tau_\infty$ es mayor que cualquier otro $\tau$, por lo que el promedio de éste es $\infty$

```julia
function dist_caminante_log(L,N,len,zoom) 
    distri=zeros(int(len/2)+2)         
    num_excesos=0                      
    for i in 1:N                       
        t=caminante(L)                 
        if t<=len && distri[int((t+1)/2)] < N/zoom           
            distri[int((t+1)/2)]+=1                    
        else                            
            num_excesos+=1              
        end
    end
    N=N-num_excesos
    distri=float(distri)/float(N)
    u=zeros(int(len/2)+2)
    for k in 1:length(u)
        u[k]=2*k-1
    end
    distri=log(distri)
    plot(u,distri)
end

for j in 5:5:50
    x=dist_caminante_log(j,1000000,1000,1)
end

#x=dist_caminante_log(10,1000000,1000,1)
#x=dist_caminante_log(100,1000000,1000,1)
```
