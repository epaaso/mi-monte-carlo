#### Inciso 1

```julia
function P_0(L) #L es la mitad de  cuanto se extiende la distribucion 
    P_cero=zeros(2*L+1)
    P_cero[L+1]=1
    return P_cero
end
```

```julia
function P_t_mas_1(P_t,p_derecha=0.5,p_stay=0.0) #p es la probabilidad de que dé un paso a la derecha
    L=length(P_t)
    P_t_mas_uno=zeros(L)
    for i in 1:L-1            #hace por separado los valores que se heredan al siguiente paso a la derecha y a la izquierda
        P_t_mas_uno[i+1]+=p_derecha*P_t[i]
        P_t_mas_uno[i]+=(1-p_derecha-p_stay)*P_t[i+1]
    end
    for j in 1:L
        P_t_mas_uno[j]+=p_stay*P_t[j]
    end
    return P_t_mas_uno
end
```

#### Inciso 2

```julia
using PyPlot
```

```julia
function P_sub(T,p=0.5,stay=0,intervalo_x=5,P_0::Function=P_0)                   
    distribuciones=Any[P_0(T)]
    for i in 1:T
        u=[-T:T]
        u=u/sqrt(i)
        push!(distribuciones,P_t_mas_1(distribuciones[i],p,stay))
        plot(u,sqrt(i)*(distribuciones[i+1]),"o-")
        xlim(-intervalo_x,intervalo_x)
    end
end

P_sub(100)
```

Observando las gráficas, queda como una distribución normal. La expresión analítica es la de la ecuación normal

## Tiempos de primer paso: sistemas con absorción

Enumeración exacta también se puede aplicar para calcular propiedades de sistemas con absorción para calcular *tiempos de primer paso* (el tiempo para llegar a un sitio *por primera vez*), tal como estudiamos en la tarea 2. Esto se lleva a cabo como sigue.

[3] Implementa enumeración exacta para un caminante aleatorio en una red 1D con una frontera reflejante en $L$ y una frontera absorbente en $0$, como sigue.

Es decir, la probabilidad que llega a $0$ en el tiempo $t$ representa la probabilidad que se *absorbe* en ese tiempo (desaparece del sistema), que corresponde a la probabilidad de que el caminante llegue a 0 *por primera vez* ("tiempo de primer paso").
Esta probabilidad se guarda en un arreglo aparte. Luego a la $P_t(0)$ correspondiente *se le asigna 0*, lo cual corresponde al hecho de que la probabilidad desaparece del sistema.

¿Cómo se satisface la conservación de probabilidad en este caso?

```julia
function caja_0(L) #funcion de condicion inicial
    caja_cero=zeros(L+1)
    caja_cero[2]=1
    return caja_cero
end
```

```julia
function caja_t_mas_1(caja_t,p_derecha=0.5,p_stay=0.0)
    L=length(caja_t)
    P_t_mas_uno=zeros(L)
    
    P_t_mas_uno[1]=(1-p_derecha-p_stay)*caja_t[2]
    P_t_mas_uno[L]=p_derecha*(caja_t[L]+caja_t[L-1])
    for i in 2:L-1      #los extremos los tratamos aparte
        P_t_mas_uno[i]=p_derecha*caja_t[i-1]+(1-p_derecha-p_stay)*caja_t[i+1]
    end
    for j in 1:L
        P_t_mas_uno[j]+=p_stay*caja_t[j]
    end
    return P_t_mas_uno
end
```

```julia
function F_t_DeCaja(L,T_max,p_derecha=0.5,p_stay=0)
    distribuciones=Any[caja_0(L)]
    dist_t=zeros(T_max+1)
    for i in 1:T_max
        push!(distribuciones,caja_t_mas_1(distribuciones[i],p_derecha,p_stay))
        dist_t[i+1]=distribuciones[i+1][1]
        distribuciones[i+1][1]=0
    end
    return dist_t
end
```

La probabilidad se conserva en 2 lugares distintos, en la distribucion de probabilidad en la caja en la distribucion de probabilidad del tiempo que tarda en llegar a 0

[4] Utiliza tu código para calcular la distribución de probabilidad $f_T$ del tiempo de primer paso al origen, $T$, para distintos tamaños $L$, y grafícalas en una sola gráfica.

```julia
function plot_muchas_cosas(L_max,T_max,p_derecha=0.5,p_stay=0)
    u=[0:T_max]
    for i in 1:L_max
        plot(u,F_t_DeCaja(i,T_max,p_derecha,p_stay),"o-")
    end
end

plot_muchas_cosas(20,100)
```
